##Sg(nT)

```python
import numpy as np
import matplotlib.pyplot as plt

phi0 = np.pi
fdpT = -0.125
dfT = 0.5
# ti = 4*T
Ni = 4

A = dfT / Ni
B = 2 * fdpT - dfT
C = -fdpT * Ni + phi0 / np.pi

print(f'A = {A}')
print(f'B = {B}')
print(f'C = {C}')

def SGnt(A, B, C, n):
  phi = []
  phi_mult_pi = []
  cos = []
  sin = []
  print(f'\n=============================Sg(nT)=====================================')
  print(f'г---т-----------------т----------------------т--------------т-------------')
  print(f'| n |A*n^2 + B*n + C  |[A*n^2 + B*n + C] * pi| cos          |sin')
  print(f'|---|-----------------|----------------------|--------------|-------------')
  for i in range(0, n):
    print(f'|{i}  |{A*i**2 + B*i + C}_____________|{np.round((np.pi * (A*i**2 + B*i + C)), 3)}_________________|{np.round(np.cos((np.pi * (A*i**2 + B*i + C))), 3)}_________|{np.round(np.sin((np.pi * (A*i**2 + B*i + C))), 3)}')
    phi.append(A*i**2 + B*i + C)
    phi_mult_pi.append((np.pi * (A*i**2 + B*i + C)))
    cos.append(np.cos((np.pi * (A*i**2 + B*i + C))))
    sin.append(np.sin((np.pi * (A*i**2 + B*i + C))))
  return phi, phi_mult_pi, cos, sin

phi, phi_mult_pi, cos, sin = SGnt(A=A, B=B, C=C, n=Ni)

print()
plt.figure(figsize=(16, 4))

plt.subplot(1, 4, 1)
plt.plot(phi)
plt.title('A*n^2 + B*n + C')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)

plt.subplot(1, 4, 2)
plt.plot(phi_mult_pi)
plt.title('[A*n^2 + B*n + C] * pi')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)


plt.subplot(1, 4, 3)
plt.plot(cos)
plt.title('cos(pi*[A*n^2 + B*n + C])')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)

plt.subplot(1, 4, 4)
plt.plot(sin)
plt.title('sin(pi*[A*n^2 + B*n + C])')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)

plt.show()

```

## ОФ(nu*T)

```python
def OF(A, B, n):
  phi = []
  phi_mult_pi = []
  cos = []
  sin = []
  print(f'\n============================= OФ(nu*T)======================================')
  print(f'г---т-----------------т----------------------т--------------т-------------')
  print(f'|n  |A*n^2 + B*n      |[A*n^2 + B*n    ] * pi| cos          |sin')
  print(f'|---|-----------------|----------------------|--------------|-------------')
  for i in range(0, n):
    print(f'|{i}  |{A*i**2 + B*i }_____________|{np.round((np.pi * (A*i**2 + B*i )), 3)}_________________|{np.round(np.cos((np.pi * (A*i**2 + B*i ))), 3)}_________|{np.round(-np.sin((np.pi * (A*i**2 + B*i ))), 3)}')
    phi.append(A*i**2 + B*i )
    phi_mult_pi.append((np.pi * (A*i**2 + B*i )))
    cos.append(np.cos((np.pi * (A*i**2 + B*i ))))
    sin.append(-np.sin((np.pi * (A*i**2 + B*i ))))
  return phi, phi_mult_pi, cos, sin

A_of = dfT / Ni
B_of = -dfT

phi_OF, phi_mult_pi_OF, cos_OF, sin_OF = OF(A_of, B_of, Ni)

print()
plt.figure(figsize=(16, 4))

plt.subplot(1, 4, 1)
plt.plot(phi_OF)
plt.title('A*n^2 + B*n + C')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)

plt.subplot(1, 4, 2)
plt.plot(phi_mult_pi_OF)
plt.title('[A*n^2 + B*n + C] * pi')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)


plt.subplot(1, 4, 3)
plt.plot(cos_OF)
plt.title('cos(pi*[A*n^2 + B*n + C])')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)

plt.subplot(1, 4, 4)
plt.plot(sin_OF)
plt.title('sin(pi*[A*n^2 + B*n + C])')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)

plt.show()
```

```python
import numpy as np
import sympy
import matplotlib.pyplot as plt

# Служебные переменные
LAST_FIGURE_NUMBER = 0

# Константы
i = complex(0, 1)
pi = np.pi

# Данные по варианту
T = 0.25
t_i = 1.25
t_p = 2
f = 1

t_0 = 0.5
fi_0 = 3 * np.pi/ 4

n_i = t_i / T
n_p = t_p / T
n_0 = t_0 / T

# Построение фигуры
def make_figure() -> plt.figure:
    """Возвращает объект для нового графика."""
    global LAST_FIGURE_NUMBER
    LAST_FIGURE_NUMBER += 1
    return plt.figure(LAST_FIGURE_NUMBER)

# Расчет бабочки в номере 3
def Butterfly(seq: list):
    """ Расчет методом бабочки фурье."""
    i = complex(0, 1)
    E = np.exp(-1 * i * (np.pi / 4))
    e_values = [E**0, E**1, E**2, E**3]
    # print(seq)

    stage_1 = [0]*8
    CalculateButterflySection(seq, stage_1, 0, 4, e_values, 0) # Размер берется в 2 раза меньше, так-как крест-накрест
    stage_1_rounded = np.real_if_close(np.around(stage_1, decimals=3), tol=10)
    # print("stage 1 results:")
    # print(stage_1_rounded)

    stage_2 = [0]*8
    e_values = [E**0, E**2, E**0, E**2]

    for i in range(0, 8, 4):
        CalculateButterflySection(stage_1, stage_2, i, 2, e_values, int(i/2))
    stage_2_rounded = np.real_if_close(np.around(stage_2, decimals=3), tol=10)
    # print("stage 2 results:")
    # print(stage_2_rounded)


    stage_3 = [0]*8
    e_values = [E ** 0, E ** 0, E ** 0, E ** 0]
    for i in range(0, 8, 2):
        CalculateButterflySection(stage_2, stage_3, i, 1, e_values, int(i/2))
    stage_3_rounded = np.real_if_close(np.around(stage_3, decimals=3), tol=10)
    # print("stage 3 results:")
    # print(stage_3_rounded)

    return stage_3_rounded

def CalculateButterflySection(base_column: list, new_column: list, section_start: int, size: int, e_values:list, cntr):
    """ Расчет секции для бабочки фурье."""
    e_counter = cntr
    for i in range(section_start, section_start + size):
        new_column[i] = base_column[i] + base_column[i + size]
        new_column[i + size] = (base_column[i] - base_column[i + size])*e_values[e_counter]
        e_counter += 1

if __name__ == '__main__':
    n = sympy.symbols('n')
    sg = sympy.cos(2*pi*f*n*T + fi_0) + i * sympy.sin(2*pi*f*n*T + fi_0)
    zeros = (0, 6, 7)
    data = []
    for i in range(8):
        if i in zeros:
            data.append(0)
        else:
            value = sg.subs(n, i)
            Re, Im = value.as_real_imag()
            data.append(complex(round(Re, 3), round(Im, 3)))

    bpf_result_no_ordered = Butterfly(data)
    order = [0, 4, 2, 6, 1, 5, 3, 7]
    bpf_result = []
    print(f"Неотсортированные данные: {bpf_result_no_ordered}")
    for order_index in order:
        bpf_result.append(bpf_result_no_ordered[order_index])
    print(f"Упорядоченные данные: {bpf_result}")

    re_values = []
    im_values = []
    amps = []

    for value in bpf_result:
        re_values.append(value.real)
        im_values.append(value.imag)
        amps.append(np.sqrt(value.real**2 + value.imag**2))
    phases = np.angle(list(map(complex, bpf_result)))

    print(f"Re values: {re_values}")
    print(f"Im values: {im_values}")
    print(f"Amps: {list(map(lambda x: round(x, 2), amps))}")
    print(f"Phases: {phases}")
    fig = make_figure()
    plt.grid()
    plt.plot(np.arange(0, 8), re_values)
    plt.title("Re part of signal")

    fig1 = make_figure()
    plt.grid()
    plt.plot(np.arange(0, 8), im_values)
    plt.title("Im part of signal")

    fig2 = make_figure()
    for i in range(8):
        plt.vlines(i, 0, amps[i], colors='r')
    plt.grid()

    fig3 = make_figure()
    for i in range(8):
        plt.vlines(i, 0, phases[i], colors='r')
    plt.grid()
    plt.yticks(np.arange(-1 * np.pi, 2 * np.pi, np.pi), [str(i) + "π" for i in range(-1, 2)])
    plt.show()
```
