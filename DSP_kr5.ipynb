##Sg(nT)

```python
import numpy as np
import matplotlib.pyplot as plt

phi0 = np.pi
fdpT = -0.125
dfT = 0.5
# ti = 4*T
Ni = 4

A = dfT / Ni
B = 2 * fdpT - dfT
C = -fdpT * Ni + phi0 / np.pi

print(f'A = {A}')
print(f'B = {B}')
print(f'C = {C}')

def SGnt(A, B, C, n):
  phi = []
  phi_mult_pi = []
  cos = []
  sin = []
  print(f'\n=============================Sg(nT)=====================================')
  print(f'г---т-----------------т----------------------т--------------т-------------')
  print(f'| n |A*n^2 + B*n + C  |[A*n^2 + B*n + C] * pi| cos          |sin')
  print(f'|---|-----------------|----------------------|--------------|-------------')
  for i in range(0, n):
    print(f'|{i}  |{A*i**2 + B*i + C}_____________|{np.round((np.pi * (A*i**2 + B*i + C)), 3)}_________________|{np.round(np.cos((np.pi * (A*i**2 + B*i + C))), 3)}_________|{np.round(np.sin((np.pi * (A*i**2 + B*i + C))), 3)}')
    phi.append(A*i**2 + B*i + C)
    phi_mult_pi.append((np.pi * (A*i**2 + B*i + C)))
    cos.append(np.cos((np.pi * (A*i**2 + B*i + C))))
    sin.append(np.sin((np.pi * (A*i**2 + B*i + C))))
  return phi, phi_mult_pi, cos, sin

phi, phi_mult_pi, cos, sin = SGnt(A=A, B=B, C=C, n=Ni)

print()
plt.figure(figsize=(16, 4))

plt.subplot(1, 4, 1)
plt.plot(phi)
plt.title('A*n^2 + B*n + C')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)

plt.subplot(1, 4, 2)
plt.plot(phi_mult_pi)
plt.title('[A*n^2 + B*n + C] * pi')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)


plt.subplot(1, 4, 3)
plt.plot(cos)
plt.title('cos(pi*[A*n^2 + B*n + C])')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)

plt.subplot(1, 4, 4)
plt.plot(sin)
plt.title('sin(pi*[A*n^2 + B*n + C])')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)

plt.show()

```

## ОФ(nu*T)

```python
def OF(A, B, n):
  phi = []
  phi_mult_pi = []
  cos = []
  sin = []
  print(f'\n============================= OФ(nu*T)======================================')
  print(f'г---т-----------------т----------------------т--------------т-------------')
  print(f'|n  |A*n^2 + B*n      |[A*n^2 + B*n    ] * pi| cos          |sin')
  print(f'|---|-----------------|----------------------|--------------|-------------')
  for i in range(0, n):
    print(f'|{i}  |{A*i**2 + B*i }_____________|{np.round((np.pi * (A*i**2 + B*i )), 3)}_________________|{np.round(np.cos((np.pi * (A*i**2 + B*i ))), 3)}_________|{np.round(-np.sin((np.pi * (A*i**2 + B*i ))), 3)}')
    phi.append(A*i**2 + B*i )
    phi_mult_pi.append((np.pi * (A*i**2 + B*i )))
    cos.append(np.cos((np.pi * (A*i**2 + B*i ))))
    sin.append(-np.sin((np.pi * (A*i**2 + B*i ))))
  return phi, phi_mult_pi, cos, sin

A_of = dfT / Ni
B_of = -dfT

phi_OF, phi_mult_pi_OF, cos_OF, sin_OF = OF(A_of, B_of, Ni)

print()
plt.figure(figsize=(16, 4))

plt.subplot(1, 4, 1)
plt.plot(phi_OF)
plt.title('A*n^2 + B*n + C')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)

plt.subplot(1, 4, 2)
plt.plot(phi_mult_pi_OF)
plt.title('[A*n^2 + B*n + C] * pi')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)


plt.subplot(1, 4, 3)
plt.plot(cos_OF)
plt.title('cos(pi*[A*n^2 + B*n + C])')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)

plt.subplot(1, 4, 4)
plt.plot(sin_OF)
plt.title('sin(pi*[A*n^2 + B*n + C])')
# plt.xlabel('n')
# plt.ylabel('')
plt.grid(True)

plt.show()
```
